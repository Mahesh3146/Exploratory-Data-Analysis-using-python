## Handling the Outliers
  1.Detecting the outliers.

  2.Treatment of outliers.

```python
# we are copying the original data to another variable

duplicate_data = data
```

```python
duplicate_data.head()
```

### IQR Method

```python
""" there are different techniques are there to detecting the outliers 
 one of them are "IQR method"

 IQR means inter quantile range 
 it will shows: 
 * lower bound of a  dataset 
 * higher bound of a dataset
"""
p0 = duplicate_data['lat'].min()
p100 = duplicate_data['lat'].max()
q1 = duplicate_data['lat'].quantile(0.25)
q2 = duplicate_data['lat'].quantile(0.5)
q3 = duplicate_data['lat'].quantile(0.75)
IQR = q3-q1
```

```python
# formulas of lower bound range and higher bound range
b = q1 - 1.5*IQR        # Lower Boundry
ub = q3 + 1.5*IQR        # Upper Boundry
```

```python
lb,ub
```

```python
print("p0 = ",p0)
print("p100 = ",p100)
print("lb = ",lb)
print("ub = ",ub)
```

```python
# we are checking is there any extreme values are there in our dataset

print(lb < p0)
print(ub > p100)
```

```python
# we are visualizing the outliers using "boxplot" chart

plt.figure(figsize=(12,9))
duplicate_data['lat'].plot(kind="box")
```

```python
duplicate_data['lat'].clip(upper=ub)
```

```python
# we are removing the extreme values and again rechecking the outliers using "boxplot"

duplicate_data['lat'].clip(upper=ub,inplace = True)
plt.figure(figsize=(12,10))
duplicate_data['lat'].plot(kind="box")
```
